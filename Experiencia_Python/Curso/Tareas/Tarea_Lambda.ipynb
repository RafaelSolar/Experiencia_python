```python
def Voc_conso(s):
    """
    Esta función regresa si el numero de vocales es mayor al numero de consonantes

    arg => s = palabra dad
    return => True si hay mas vocales y flase si hay mas consonates
    """

    s=s.lower()
    ls=s.split(" ")
    sf=ls[0]
    numvoc=sf.count("a")+sf.count("e")+sf.count("i")+sf.count("o")+sf.count("u")
    numcons=len(sf)-numvoc

    return numvoc > numcons
words=["Ave", "Paloma", "Tarea", "Compartir"]

list(filter(Voc_conso, words))
```

```python
def divisoresn(n):
    """
    Funcion que obtiene los divisores de un entero positivo
    arg: n - numero entero positivo
    return: lista de divisores

    """
    divisors=[]
    n=int(n)
    for i in range(1,n+1):
        if n % i == 0:
            divisors.append(i)
    return divisors


nums=[48,69,5,21,99,897,354,13]
list(filter(lambda l : len(divisoresn(l)) >5, nums))
```

```python
def longword(s1,s2):
    """
    Funcion que devuelve cual es la palabra más larga

    args => s1 = palabra 1
            s2 = palabra 2
    
    return => la palabra más arga
    """
    s1, s2 = s1.lower(), s2.lower()
    ls1, ls2=s1.split(" "), s2.split(" ")
    s1f, s2f = ls1[0], ls2[0]
    
    if len(s1f) > len(s2f):
        return s1f
    elif len(s1f) < len(s2f):
        return s2f
    else:
        return [s1f,s2f]


from functools import reduce
words=["Ave", "Paloma", "Tarea", "Compartir","Manzanas"]
reduce(longword,words)

```

```python
def Num_bigger(n,m):
    """
    Devuelve el numero mayor
    args => n: numero uno
            m: numero dos
    retunr => el numero mayor

    """
    if n > m:
        return n
    elif n < m:
        return m
    else:
        return [n,m]



def Num_Voc(s):
    """
    Esta función regresa  el numero de vocales 

    arg => s = palabra dad
    return => El numero de vocales
    """

    s=s.lower()
    ls=s.split(" ")
    sf=ls[0]
    numvoc=sf.count("a")+sf.count("e")+sf.count("i")+sf.count("o")+sf.count("u")

    return numvoc

Num_Cons= lambda s: len(s) - Num_Voc(s)

words=["Murcielago","Editorial","Ave", "Paloma", "Tarea"]

words_cons=list(map(Num_Cons, words))
word_mascons = reduce(Num_bigger, words_cons)
pos=-1
for i in words_cons:
    pos +=1
    if i == word_mascons:
        break
    else:
        continue
print("La palabra con más consonantes es {}".format(words[pos]))
```
