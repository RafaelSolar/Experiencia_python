```python
n=int(input("Introduce el numero de filas (matriz nxn): "))
A=np.zeros((n,n))
for i in range(n):
    for j in range(n):
        if i==j:
            A[i,j]=1
        else:
            A[i,j]=0
print("Matriz identidad de dimention {}x{}".format(n,n))
print(A)
```
