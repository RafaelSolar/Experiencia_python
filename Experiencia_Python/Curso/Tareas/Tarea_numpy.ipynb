```python
def confpalindromo(s):
    """
    Funcion que comprueba si una palabra es un palindromo
    arg: s - palabra 
    return: True si es un palindromo, false si no es palindromo
    
    """ 
    s=s.lower()
    lword, lwordrev = [], [] 
    for i in range(0,len(s)):
        lword.append(s[i])
        lwordrev.append(s[(i*-1)-1])
    resl=lword == lwordrev

    return resl


words = np.array([["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato",
"elle"]])

ispalindromo = np.frompyfunc(confpalindromo, 1,1)
ispalindromo(words)
```

```python
def acer(am,n):
    amn = []
    for i in range(0,5):
        if am[i] in amn:
            continue
        else:
            amn.append(am[i])
    amn = np.array(amn)
    count = 0
    for e in amn:
        if e in n:
            if len( n[n == e]) >= len( am[am == e]):
                count += len(am[am == e])
            else:
                count += len(n[n == e])
    return count



n = rd.randint(1,21, size=5)
print(n)
a = []
for _ in range(0,5):
    ae = int(input("Introduce un numeroentre el 1-20: "))
    a.append(ae)
am = np.array(a)
print(am)



wcom = [len(np.where(n == am)[0]), acer(am,n) ]

while wcom != [5,5]:
    print("Vuelve a intentarlo")
    print("Numeros acertados: {}\nNumero bien colocados: {}".format( wcom[1], wcom[0] ))
    a = []
    for _ in range(0,5):
        ae = int(input("Introduce un numeroentre el 1-20: "))
        a.append(ae)
    am = np.array(a)
    print(am)
    wcom = [len(np.where(n == am)[0]), acer(am,n)]
else:
    print("¡¡¡ENHORABUENA!!! Has acertado el array que había pensado: ",n )

```

```python
def numvoc(s):
    s=s.lower()
    ls=s.split(" ")
    sf=ls[0]
    return sf.count("a")+sf.count("e")+sf.count("i")+sf.count("o")+sf.count("u")
num_voc = np.frompyfunc(numvoc,1,1)
words = np.array([["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato",
"elle"]])
print(num_voc(words))
```

```python
def str_in_set(s,conj):
    return s in conj

str_in_set_array = np.frompyfunc(str_in_set,2,1)
words = np.array([["avion", "gato", "arroz"], ["cena", "kayak", "raton"], ["pluma", "perro",
"ellos"]])
conj = {"ala", "delfin", "arroz","cena", "kayak", "picnic","hoja", "gato","elle"}
print(str_in_set_array(words,conj))
```
