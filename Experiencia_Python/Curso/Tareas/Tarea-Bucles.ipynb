```python
s=input("Introduce una palabra: ")
l=input("Introduce una letra: ")
s=s.lower()
l=l.lower()
while s.find(" ")!=-1 or len(l) !=1:
    if s.find(" ")!=-1:
        s=input("Introduce una palabra: ")
        s=s.lower()
    elif len(l) !=1:
        l=input("Introduce una letra: ")
        l=l.lower()

count=0
for c in s:
    if c == l:
        count +=1

if count>0:
    print("La palabra ({}) contiene {} vez la letra {}.".format(s,count,l))
else:
    print("La palabra ({}) no contiene la letra {}.".format(s,l))

```

```python
print("Programa para calcular presupuesto, cuando indique un 0 significa que desea terminar el programa.")
sum=0.0
while sum<=200.0:
    pres=float(input("Introduce la cantidad: "))
    if pres==0.0:
        print()
        print("La cantidad final es {} $".format(sum))
        break
    else:
        sum +=pres
else:
    print("Se ha pasado del presupuesto de 200$, ya que el resultado es {}".format(sum))
```

```python
num=int(input("Introduce un numero entero: "))

countpos=0
countneg=0
while num!=0 :
    if num>0:
        countpos +=1
    elif num<0:
        countneg +=1
    num=int(input("Introduce un numero entero: "))
else:
    if countpos!=0 or countneg !=0:
        print("El número de valores positivos dados es {}.\nEl número de valores negativos dados es {}.".format(countpos,countneg))
    else:
        print("No se pudo guardar ningún valor")
```

```python
print("Programa para calcular la media, cuando indique un 0 significa que desea terminar el programa.")
sum=0.0
count=0
num=float(input("Introduce un valor: "))
while num!=0.0:
    sum +=num
    count+=1
    num=float(input("Introduce un valor: "))
else:
    if count==0:
        print("No introdujo ningún valor.")
    else:
        print("La media aritmetica es = {}".format(sum/count))
    
```

```python
cantidad=int(input("Cuantos numeros (más que 1) va a introducir: "))
res=0.0
fin=0
if cantidad>1:
    for i in range(cantidad+1):
        num=float(input("Indroduzca un número: "))
        if res==0.0:
            res=num
        else:
            res *=num
            fin+=1
            if fin == cantidad-1:
                print("El producto de los numeros es {}".format(res))
                break

else:
    print("No introdujo ningún número.")

```
