```python
def RGB_to_HEX(r = 0,g = 0,b = 0):
    r,g,b = str(hex(r)),str(hex(g)),str(hex(b))
    R,G,B = r[2:4],g[2:4],b[2:4]
    colorhex = "#{}{}{}".format(R,G,B)
    return colorhex
```

# Problema 4

```python
def funciones(fun = 0, x=0):
    if fun == 0:
        y = ((mt.sin(x))/x)
    elif fun == 1:
        y = ( (2*(mt.sin(x)))/x )
    else:
        y = -((mt.sin(x))/x)
    return y

colores = [RGB_to_HEX(240,103,65), RGB_to_HEX(164,240,65),RGB_to_HEX(141,65,240)]
rango_x = np.linspace(-5*mt.pi,5*mt.pi, num = 100)
plt.figure(figsize=(8,6))
for j in range(0,3):
    plt.plot(rango_x, [funciones(j,i) for i in rango_x], c = colores[j])
plt.title("Funciones")
plt.xlabel("x")
plt.ylabel("y")

plt.show
```

# Problema 5

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.title("sen(x)/x")
plt.ylabel("y")
plt.xlabel("x")
plt.ylim((-1.2,2.2))
plt.plot(rango_x,list(map(lambda X : mt.sin(X)/X,rango_x)))
plt.subplot(1,3,2)
plt.title("2sen(x)/x")
plt.ylabel("y")
plt.xlabel("x")
plt.ylim((-1.2,2.2))
plt.plot(rango_x,list(map(lambda X : (2*mt.sin(X))/X,rango_x)))
plt.subplot(1,3,3)
plt.title("-sen(x)/x")
plt.ylabel("y")
plt.xlabel("x")
plt.ylim((-1.2,2.2))
plt.plot(rango_x,list(map(lambda X : -mt.sin(X)/X,rango_x)))

plt.suptitle("Funciones-subplot")
```
