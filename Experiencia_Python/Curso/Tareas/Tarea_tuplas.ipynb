```python
num=int(input("Cuantos numeros vas a introducir: "))
li=[]
for _ in range(num):
    x=int(input("Numero: "))
    wpi=""
    if x%2 ==0:
        wpi="par"
    else:
        wpi="impar"
    li.append((x,wpi))
print("Numero--Par/impar")
for e in li:
    print(e)
```

```python
s=str(input("Introduce una frase: "))
ls=s.split(" ")
listtu=[]
pos=0
for e in ls:
    for i in range(len(ls)):
        word=(e,len(e),e[0],pos+1)
    listtu.append(word)
    pos+=1
print("Palabra--Longitud--Inicial--Posicion")
for tu in listtu:
    print(tu)


```

```python
s=str(input("Introduce una frase: "))
lwords=[]
while s!="":
    lwords.append(s)
    s=str(input("Introduce una frase: "))

if len(lwords)>0:
    tuword=tuple(lwords)
    (firstw, *_ , lastw)=tuword
    print("Esta es la primer palabra {}.\nEsta es la segunda palabra {}.".format(firstw,lastw))
else:
    print("No se ha escrito ninguna palabra.")
```

```python
lw,llon=[],[]
s=str(input("Introduce una palabra: "))
while s!="":
    lw.append(s)
    llon.append(len(s))
    s=str(input("Introduce una frase: "))
if len(lw)>0:
    tuword, tulon=tuple(lw),tuple(llon)
    diccwl=dict(zip(tuword,tulon))
    for key,value in diccwl.items():
        print("Palabra {}, longitud {}".format(key,value))
else:
    print("No se ha escrito ninguna palabra.")
```

```python
count,lim=0,20
lnum,lmul,lista=[],[],[]
while count>=0 and count<lim:
    x=int(input("Introduce un numero entre 1 y 20: "))
    if x>=1 and x<=20:
        lnum.append(x)
        count+=1
    else:
        continue
for i in lnum:
    lmul=list(range(i,(i*10)+1,i))
    lista.append((i,lmul))


for tup in lista:
    print(tup)

for i in range(1,21):
    for j in range(11):
        print("{} x {} = {}",format(i,j,i*j))
```

```python
x=int(input("Introduce un numero entero: "))
y=int(input("Introduce otro entero: "))

if x>=y:
    tup=(x,y,x//y,x%y)
    print("Dividendo-Divisor-Cociente-Resto")
    print(tup)
else:
    print("El numero {} es menor a {}".format(x,y))
```
