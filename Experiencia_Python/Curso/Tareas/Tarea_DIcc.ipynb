```python
dicc_compra, sum={},0.
print("Presione 0 en articulo para terminar la lista de compras.")
art=input("Introuduce el articulo: ")
while art!="0":
    dicc_compra[art]=float(input("Precio del producto: "))
    sum += dicc_compra[art]
    art=input("Introuduce el articulo: ")
else:
    dicc_compra["Total"]=sum
    for key, value in dicc_compra.items():
        print("{} : {}" .format(key,value))
```

#9

```python
num_word=int(input("Numero de palabras que introducirá= "))
dicc_word={}
for _ in range(num_word):
    voc=0
    word=input("Introduce la palabra: ")
    for c in word.lower():
        if c=="a" or c=="e" or c=="i" or c=="o" or c=="u":
            voc+=1
        else:
            continue
    dicc_word[word]=voc
for key, value in dicc_word.items():
    print("Palabra==N_vocales")
    print(key , ":",value)
```

```python
import numpy as np
n=int(input("Introduce el numero de filas: "))
m=int(input("Introduce el numero de columnas: "))
A=np.empty((n,m))


for i in range(n):
    for j in range(m):
        A[i,j]=float(input("Introduce el valor de la posicion ({},{}): ".format(i+1,j+1)))
dicc_filas={}
i=1
for row in A:
    dicc_filas["Fila {}".format(i)]=row
    i+=1

dicc_matrix={"N_Filas":n,"N_Columnas":m,"Elementos_Filas":dicc_filas}
for key,value in dicc_matrix.items():
    print(key, ":",value)
```
