```python
print("El primer número debe ser positivo y mayor o igual al segundo, mientras que el segundo debe ser positivo sin incluir al cero.\n")
x=int(input("Introduzca el primer número: "))
y=int(input("Introduzca el segundo número: "))
if x>0 and y>0 and x>=y and y!=0:
    if x%y==0:
        print("La division es entera y su cociente es {}".format(x//y))
    else:
        print("La división no es entera puesto que el cociente es {} y el resto es {}".format(x//y,x%y))
else:
    print("Los numeros no cumplen con las condiciones. Try again")
```

```python
x=int(input("Introduzca el primer número(distinto a cero y positivo): "))
y=int(input("Introduzca el segundo número(distinto a cero y positvo): "))
if x>0 and y>0:
    if x>y:
        print("El primer numero ({}) es multiplo de ({}).".format(x,y)) if x%y==0 else print("El primer numero ({}) no es multiplo de ({}).".format(x,y))
    elif x<y:
        print("El segundo numero ({}) es multiplo de ({}).".format(y,x)) if y%x==0 else print("El segundo numero ({}) no es multiplo de ({}).".format(y,x))
    else:
        print("Los numeros son iguales.")
else:
    print("Los numeros no cumplen con las condiciones. Try again")
```

```python
letra=input("Introduce una letra: ").lower()
if len(letra)==1:
    if letra=="a":
        print("La letra es a")
    elif letra=="e":
        print("La letra es e")
    elif letra=="i":
        print("La letra es i")
    elif letra=="o":
        print("La letra es o")
    elif letra=="u":
        print("La letra es u")
    else:
        print("La letra no es vocal.")
else:
    print("No introdujo bien el dato. Try again")
```

```python
from math import sqrt as raiz
a=float(input("Introduzca a (distinto a 0): "))
b=float(input("Introduzca b: "))
c=float(input("Introduzca c: "))
dis=b**2-(4*a*c)
if a!=0:
    if dis>0:
        x1=(-b+raiz(dis))/(2*a)
        x2=(-b-raiz(dis))/(2*a)
        print("x1 = {}\nx2 = {}".format(x1,x2))
    elif dis<0:
        print("No contiene raices en los numeros reales")
    else:
        print("Las dos soluciones son iguales, por lo tanto:\nx = {}".format(-b/(2*a)))
else:
    print("a es igual a cero, por lo tanto, no se puede resolver.")
```
