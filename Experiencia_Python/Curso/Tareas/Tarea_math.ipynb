```python
def degree_to_rad(x = 0.):
    from math import pi
    mult = x//360.
    if mult>0:
        return ((x-360*mult)*pi)/180
    return (x*pi)/180

def rad_to_degree(x =0.):
    from math import tau
    mult = x // tau
    if mult > 0:
        return ((x-mult*tau)*180)/(tau/2.)
    return (x*180)/(tau/2.)

print("Comprovacion de grados a radianes:\nFuncion creada : {} , Funcion math: {}".format(degree_to_rad(450), mt.radians(90)))
print()
print("Comprovacion de radianes a grados:\nFuncion creada : {} , Funcion math: {}".format(rad_to_degree(3*mt.pi), mt.degrees(mt.pi)))
```

```python
def choose(n, k):
  import math
  if (n >= k and k >= 0):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
  else:
    return "No se puede calcular el n√∫mero factorial indicado"

choose(15,3)
```

```python
z = complex(1 , 8.5)
ez = cmt.exp(z)
ezm = cmt.exp(-z) 
print("senh({}) = {} , (e^z - e^-z)/2 = {}".format(z,cmt.sinh(z) , (ez - ezm)/2))
print()
print("cosh({}) = {} , (e^z + e^-z)/2 = {}".format(z,cmt.cosh(z) , (ez + ezm)/2))
```
