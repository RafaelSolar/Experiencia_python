```python
def divisoresn(n):
    """
    Funcion que obtiene los divisores de un entero positivo
    arg: n - numero entero positivo
    return: lista de divisores

    """
    divisors=[]
    n=int(n)
    if n<0:
        print("El numero {} es negativo".format(n))
        return
    elif n==0:
        print("El numero es cero, por lo tanto tiene infinitos divisores")
        return
    for i in range(1,n+1):
        if n % i == 0:
            divisors.append(i)
    
    print("Estos son los divisores: ")
    return divisors
divisoresn(245)
```

```python
def divisoresn(n):
    """
    Funcion que obtiene los divisores de un entero positivo
    arg: n - numero entero positivo
    return: lista de divisores

    """
    divisors=[]
    for i in range(1,n+1):
        if n % i == 0:
            divisors.append(i)
    return divisors

def calcularMCD(n,m):
    """
    Funcion que obtiene el maximo comun divisor (MCD)
    arg: n - numero entero positivo
         m - numero entero positivo
    return: MCD

    """
    if (n <= 0 or m <=0):
        print("Alguno numero o ambos son negativos o cero")
        return

    n , m= int(n) , int(m)
    divisorn=set(divisoresn(n))
    divisorm=set(divisoresn(m))
    interseccion= list(divisorn & divisorm)

    max=-9999

    for i in interseccion:
        if i>max:
            max=i
        else:
            continue
    
    return max

n, m= 1032, 180
MCD = calcularMCD(n,m)

print("El maximo comun denominador de {} y {} es: {}".format(n,m,MCD))
```

```python
def calcularMCM(n,m):
    """
    Funcion que obtiene el minimo comun multiplo (MCM)
    arg: n - numero entero positivo
         m - numero entero positivo
    return: MCM

    """   
    if (n <= 0 or m <=0):
        print("Alguno numero o ambos son negativos o cero")
        return
    n , m= int(n) , int(m)
    MCD = calcularMCD(n,m)
    MCM = (n*m)/MCD

    return MCM

n, m= 1032, 180
MCM = calcularMCM(n,m)
print("El minimo comun multiplo de {} y {} es: {}".format(n,m,MCM))
```

```python
def confpalindromo(s):
    """
    Funcion que comprueba si una palabra es un palindromo
    arg: s - palabra 
    return: True si es un palindromo, false si no es palindromo
    
    """ 
    s=s.lower()
    lword, lwordrev = [], [] 
    for i in range(0,len(s)):
        lword.append(s[i])
        lwordrev.append(s[(i*-1)-1])
    resl=lword == lwordrev

    return resl
    
confpalindromo("Malayalam")
```

```python
def HEXtoRGB(hexnum):

    red, green, blue = hexnum[0:2], hexnum[2:4], hexnum[4:6]

    red, green, blue = int(red,16), int(green,16), int(blue,16)

    return [red, green, blue]
    
HEXtoRGB("ff48F1")
```
