```python
s=str(input("Introduce una frase: "))
s=s.lower()
letter=input("Introduce una letra: ")
slista=s.split(" ")
words=set()
for e in slista:
    if e[0] == letter.lower():
        words.add(e)

print("Las palabras que empiezan con {} son : {}".format(letter,words))

```

```python
s=str(input("Introduce una frase: "))
s=s.lower()
letter=input("Introduce una letra: ")
letter=letter.lower()
conj=set()
slista=s.split(" ")
for word in slista:
    for i in range(len(word)):
        if word[i] == letter:
            conj.add(word)

print("La palabras que contienen la letra {} son : {}".format(letter,conj))
```

#10

```python
s=str(input("Introduce una frase: "))
s=s.lower()
conj=set()
slista=s.split(" ")
for word in slista:
    l1,l1r=[],[]
    for c in word:
        l1.append(c)
    l1.reverse()
   
    for tl in l1:
        l1r.append(tl)
    l1.reverse()
    if l1==l1r and len(l1)!=1:
        conj.add(word)

print("Los palindromos de la frase son : {}".format(conj))
```
